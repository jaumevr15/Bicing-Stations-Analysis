```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import folium
import json
```

```python
map = folium.Map(location=[41.378798,2.1600506], tiles='Stamen Watercolor', zoom_start=16)
with open('..//maps//barris.json') as json_data:
    barris = json.load(json_data)
barris_data = pd.read_csv('..//databases/50_density_df.csv', sep=',',encoding='latin-1')
#original dataset from Barcelona Open Data
```

```python
folium.Circle(
    radius=200,
    location=[41.378798,2.1600506],
    popup='Zona Sant Antoni',
    color='crimson',
    fill=True,
).add_to(map)
```

```python
folium.Choropleth(
    geo_data=barris,
    name='choropleth',
    data=barris_data,
    key_on='neighbourhood',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
).add_to(map)

folium.LayerControl().add_to(map)
```
