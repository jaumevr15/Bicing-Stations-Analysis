```python
# Load thd page and get it into Beautiful Soup
html_doc = 'https://opendata-ajuntament.barcelona.cat/data/en/dataset/estat-estacions-bicing'
page = requests.get(html_doc)
#soup = BeautifulSoup(html_doc, 'html.parser')
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#identify the file names from the OpenDataBarcelona (ODB) website
df_file_names_odb = pd.DataFrame(soup.findAll("a", {"title" : lambda L: L and L.startswith('2019_')}))
```

```python
df_file_names_odb
```

```python
bic_months = []
f_name = []

for j in range(10):
    f_name = []
    for i in range(5,len(df_file_names_odb.iloc[j][0])):
        f_name.append(df_file_names_odb.iloc[j][0][i])

    string_fname = ''.join(f_name)
    bic_months.append(string_fname)


bic_months
```

```python
# This section will now download the above named files systematically
url = 'https://opendata-ajuntament.barcelona.cat/resources/bcn/BicingBCN/{}'
  
folder = '../Dataset/mar-dec2019/'
count = 0
remain = len(bic_months)

print("This will take some time - however progress will be displayed below")
for i in bic_months:
    wget.download(url.format(i), folder)

    count += 1
    remain -= 1
    print(f'You have downloaded {count} file(s) just {remain} to go.')

print("All files downloaded. You may proceed")
```

```python
new_FN = '../Dataset/mar-dec2019/{}'

for i in bic_months:
    Archive(new_FN.format(i)).extractall('../Dataset/mar-dec2019/')
```

```python
CSV_Files = [file for file in listdir('../Dataset/mar-dec2019/') if file.endswith('.csv')]    
print(len(CSV_Files))
```

```python
CSV_Files
```

```python
df_dict = {}

dfname = 'df{}'

for i in range(0,len(CSV_Files)):    
    CSV_name = '../Dataset/mar-dec2019/'+CSV_Files[i]
    print(CSV_name)
    df_dict[dfname.format(i)] = pd.read_csv(CSV_name)
    df_dict[dfname.format(i)]['import_dt'] = CSV_name 
print("Load of zip files finished. You may proceed")
```

```python
df_staging = []
for key in df_dict:
    df_staging.append(df_dict[key])
    
df = pd.concat(df_staging, ignore_index=True, sort=True)
```
